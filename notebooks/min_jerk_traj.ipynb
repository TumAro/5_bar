# Minimum Jerk Trajectory

For our system we need the following boundary Values are,
- `pos` 0 to 1 (normalise)
- `vel` start = end = 0
- `acc` start = end = 0

Minimum jerk means minimizing $\int \frac{d^3x}{{dt}^3}^2 dt$. Calculus of variation on the system gives 1uintic polynomial. Therefore we take the following general equation of a polynomial

$$
\begin{align*}
s(t) &= a_0 + a_1 t + a_2 t^2 + a_3 t^3 + a_4 t^4 + a_5 t^5 \\
\dot{s}(t) &= a_1 + 2a_2 t + 3a_3 t^2 + 4a_4 t^3 + 5a_5 t^4 \\
\ddot{s}(t) &= 2a_2 + 6a_3 t + 12a_4 t^2 + 20a_5 t^3 \\
\end{align*}
$$

putting in all the boundary values yields the following system of equation:
$$
\begin{align*}
a_0 = a_1 = a_2 &= 0 \\
a_3 t^3 + a_4 t^4 + a_5 t^5 &= 1 \\
3a_3 t^2 + 4a_4 t^3 + 5a_5 t^4 &= 0 \\
6a_3 t + 12a_4 t^2 + 20a_5 t^3 &= 0
\end{align*}
$$
which is 3 equation with 3 unknows, it can be rewritten as the following

$$
\begin{align*}
T&C = S \\
\text{or, } &C = T^{-1}S
\end{align*}
$$
where $T$ is the $3\times3$ matrix of the power of $t$ along the numeral coefficients. $C$ is the $3\times1$ vector of the unknown variables and $S$ is the $3\times1$ vector of the solution on the R.H.S.

```python
t = Symbol('t', positive=True)
a3, a4, a5 = symbols('a3 a4 a5')

T = Matrix([
    [t**3,   t**4,    t**5],
    [3*t**2, 4*t**3,  5*t**4],
    [6*t,    12*t**2, 20*t**3]
])
R = Matrix([1,0,0])

S = T.inv() @ R
```

$t$ is in normalised form so subbing $t=1$

```python
S.subs(t,1)
```

therefore the final polynomial becomes,
$$
\begin{align*}
s(t) &= 10t^3 -15t^4 +6t^5 \\
\dot{s}(t) &= 30t^2 -60t^3 +30t^4 \\
\ddot{s}(t) &= 60t - 180t^2 + 120t^3
\end{align*}
$$
