```python
import tomllib
with open("../config.toml", "rb") as f:
    config = tomllib.load(f)

l = config["links"]["PROXIMAL"]
r = config["links"]["DISTAL"]
b = config["links"]["BASE"]
```

## Test of invertibility

`fk` and `ik` functions are already defined, now will be tested if it satisfies `fk(ik(x)) = x` for sampled points.
Sampled points would not be truly random but from the defines workspace bound of the system

```python
# Varifiying if the points are reachable by system
d_min = abs(l-r)
d_max = l+r

def verify_pts(x,y):
    d_L = sqrt(x**2 + y**2)
    d_R = sqrt((x-b)**2 + y**2)

    if d_min <= d_L <= d_max and d_min <= d_R <= d_max:
        return True
    return False

from random import uniform
points = [(uniform(0,b), uniform(0,l+r)) for _ in range(60)]
sample_pts = [pt for pt in points if verify_pts(*pt)]
len(sample_pts)
```

```python
from src import forward, inverse

inverse_pts = [inverse(*pt) for pt in sample_pts]
forward_pts = [forward(*pt) for pt in inverse_pts]
```

```python
eps = 1e-6
results = []
for orig, fwd in zip(sample_pts, forward_pts):
    match = any(abs(orig[0]-sol[0]) < eps and abs(orig[1]-sol[1]) < eps for sol in fwd)
    results.append(match)

print(f"{sum(results)}/{len(sample_pts)} passed")
```

All test passing consequences to a perfectly working numeric `forward` and `inverse` kinematic functions.
